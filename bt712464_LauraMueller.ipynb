# ***Covid ICU Dataset Analysis***


## General Overview

```python
crp = pd.read_csv('CRP.csv')
crp = pd.DataFrame(crp)
crp.head()

#c-reactive protein test checks for inflammation in the body. Inflammation can be caused by infection, injury, or chronic disease.
```

```python
hco3 = pd.read_csv('HCO^-3.csv')
hco3 = pd.DataFrame(hco3)
hco3.head()

#bicarbonate testing is part of an electrolysis panel to identify or monitor electrolyte or pH imbalance and to screen for ketosis or acidosis
```

```python
lac = pd.read_csv('Lactate.csv')
lac = pd.DataFrame(lac)
lac.head()

#The lactate blood test is primarily ordered to help determine if someone has lactic acidosis, a level of lactate that is high enough to disrupt a person's acid-base (pH) balance. Lactic acidosis is most commonly caused by an inadequate amount of oxygen in cells and tissues (hypoxia)
```

```python
spo2 = pd.read_csv('Oxygen saturation.csv')
spo2 = pd.DataFrame(spo2)
spo2.head()

#Measured with a pulsoximeter, typically attached to finger or earlobe. Normal oxygen saturation levels range between 94%-99%. When SPo2 levels fall below 93% it is a sign that oxygen therapy is required.
```

```python
pfr = pd.read_csv('P-F Ratio.csv')
pfr = pd.DataFrame(pfr)
pfr.head()

#PaO2/FiO2 (=P/F) ratio is the ratio of arterial oxygen partial pressure (PaO2 in mmHg) to fractional inspired oxygen. A P/F Ratio less than 300 indicates acute respiratory failure.
```

```python
paco2 = pd.read_csv('PaCO_2.csv')
paco2 = pd.DataFrame(paco2)
paco2.head()

#An elevated PaCO2 reflects alveolar hypoventilation, whereas a decreased PaCO2 reflects alveolar hyperventilation. Acute changes in PaCO2 will alter the pH. As a general rule, a low pH with a high PaCO2 suggests a respiratory acidosis, while a low pH with a low PaCO2 suggests a metabolic acidosis
```

```python
ppress = pd.read_csv('Peak pressure.csv')
ppress = pd.DataFrame(ppress)
ppress.head()

#Peak pressure is the pressure achieved during inspiration when the air is being pushed into the lungs and is a measure of airway resistance.
```

```python
ph = pd.read_csv('pH.csv')
ph = pd.DataFrame(ph)
ph.head()

#In the absence of pathological states, the pH of the human body ranges between 7.35 to 7.45, with the average at 7.40. Deviations indicate acidosis or
```

```python
plat = pd.read_csv('Platelet Count.csv')
plat = pd.DataFrame(plat)
plat.head()

#Thrombocytopenia is a frequent finding in intensive care unit especially among adults and medical ICU patients. Thrombocytopenia is defined as a platelet count less than 100×109/l in ICU setting.
```

```python
ventr = pd.read_csv('Ventilatory Ratio.csv')
ventr = pd.DataFrame(ventr)
ventr.head()

#The ventilatory ratio is a simple bedside index that can be calculated using routinely measured respiratory variables and is a measure of impaired ventilation. Ventilatory ratio is defined as [minute ventilation (ml/min) × PaCO2 (mm Hg)]/(predicted body weight × 100 × 37.5).
```

```python
outcome = pd.read_csv('outcome.csv')
outcome = pd.DataFrame(outcome)
outcome.head()

#Patient Outcomes 7 days post admission
```
